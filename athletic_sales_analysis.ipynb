```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

# Put each file into a separate DataFrame
df2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
df2021 = pd.read_csv('Resources/athletic_sales_2021.csv')

```

```python
# Display the 2020 sales DataFrame
df2020.head()
```

```python
# Display the 2021 sales DataFrame
df2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.

print("DataFrame 2020 info:")
df2020.columns.tolist() 
df2020.dtypes
df2020.info

```

```python
# Check the 2021 sales data types.

print("DataFrame 2021 info:")
df2021.columns.tolist()
df2021.dtypes
df2021.info
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

combined_df = pd.concat([df2020, df2021], axis=0)

# Reset the index to have sequential numbering
combined_df = combined_df.reset_index(drop=True)

# View the result
combined_df.head()
combined_df.tail()
```

```python
# Check if any values are null.

null_check = combined_df.isnull().sum()
print("Count of null values in each column:")
print(null_check)
```

```python
# Check the data type of each column
print("Data types of each column:")
combined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'])

# Format the display as YYYY-MM-DD 
combined_df['invoice_date_formatted'] = combined_df['invoice_date'].dt.strftime('%Y-%m-%d')


# Verify the result
print(combined_df['invoice_date'].head())

```

```python
# Confirm that the "invoice_date" data type has been changed.
print(combined_df['invoice_date'].dtype)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.

# Group by region, state, city and sum the units sold column
products_by_location = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the sum column
products_by_location = products_by_location.rename(columns={'units_sold': 'Total_Products_Sold'})

# Sort in descending order to see highest sales first
products_by_location = products_by_location.sort_values('Total_Products_Sold', ascending=False)

# Show the top 5 results
top_5_locations = products_by_location.head(5)

# Display the results
top_5_locations.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
# Create a pivot table

pivot_products = pd.pivot_table(
    combined_df,
    values='units_sold',
    index=['region', 'state', 'city'],
    aggfunc='sum'
)

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_products.columns = ['Total_Products_Sold']
pivot_products = pivot_products.reset_index()

# Sort in descending order
pivot_products = pivot_products.sort_values('Total_Products_Sold', ascending=False)

# Show the top 5 results
top_5_pivot = pivot_products.head(5)

# Display the results
top_5_locations.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.

# Group by region, state, city and sum the total_sales
sales_by_location = combined_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the column from 'total_sales' to 'Total Sales'
sales_by_location = sales_by_location.rename(columns={'total_sales': 'Total Sales'})

# Sort in descending order
sales_by_location = sales_by_location.sort_values('Total Sales', ascending=False)

# Show the top 5 results
top_5_sales = sales_by_location.head(5)

# Display the results
top_5_sales.head()

```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

# Create a pivot table for total sales by location
pivot_sales = pd.pivot_table(
    combined_df,
    values='total_sales',     
    index=['region', 'state', 'city'],
    aggfunc='sum'           
)

# Rename the column to "Total Sales"
pivot_sales.columns = ['Total Sales']

# Reset the index to get a standard DataFrame
pivot_sales = pivot_sales.reset_index()

# Sort in descending order to see highest sales first
pivot_sales = pivot_sales.sort_values('Total Sales', ascending=False)

# Show the top 5 results
top_5_sales_pivot = pivot_sales.head(5)

# Display the results
top_5_sales_pivot.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.

# Group by retailer and sum the total_sales column
retailer_sales = combined_df.groupby('retailer')['total_sales'].sum().reset_index()

# Rename the sum column
retailer_sales = retailer_sales.rename(columns={'total_sales': 'Total Sales'})

# Sort in descending order
retailer_sales = retailer_sales.sort_values('Total Sales', ascending=False)

# Get the retailer with the most sales (top row after sorting)
top_retailer = retailer_sales.iloc[0]

# Print the result
print(f"The retailer with the most sales is {top_retailer['retailer']} with ${top_retailer['Total Sales']:,.2f} in sales")

# Display the full ranking
print("\nFull retailer sales ranking:")
print(retailer_sales)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.

# Create a pivot table of sales by retailer
pivot_retailer_sales = pd.pivot_table(
    combined_df,
    values='total_sales',    # Column containing sales amounts
    index=['retailer'],      # Retailer as the index
    aggfunc='sum'            # Sum the sales
)

# Rename the column to "Total Sales"
pivot_retailer_sales.columns = ['Total Sales']

# Reset the index 
pivot_retailer_sales = pivot_retailer_sales.reset_index()

# Sort in descending order
pivot_retailer_sales = pivot_retailer_sales.sort_values('Total Sales', ascending=False)

# Get the top retailer
top_retailer = pivot_retailer_sales.iloc[0]

# Print the result
print(f"The retailer with the most sales is {top_retailer['retailer']} with ${top_retailer['Total Sales']:,.2f} in sales")

# Display the full ranking
print("\nFull retailer sales ranking:")
print(pivot_retailer_sales)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.

womens_athletic_footwear = combined_df[combined_df['product'] == "Women's Athletic Footwear"]
print(f"Number of Women's Athletic Footwear sales: {len(womens_athletic_footwear)}")

# Use groupby to determine which retailer sold the most
retailer_sales = womens_athletic_footwear.groupby('retailer')['total_sales'].sum().reset_index()
retailer_sales = retailer_sales.rename(columns={'total_sales': 'Total Sales'})
retailer_sales = retailer_sales.sort_values('Total Sales', ascending=False)

# Get the top retailer
top_retailer = retailer_sales.iloc[0]
print(f"\nUsing groupby: The retailer with the most Women's Athletic Footwear sales is {top_retailer['retailer']} with ${top_retailer['Total Sales']:,.2f}")

# Preview the filtered data
print("\nSample of women's athletic footwear sales:")
womens_athletic_footwear.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

# Group by retailer, region, state, city and sum the quantity

units_by_location = womens_athletic_footwear.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the sum column to 'Womens_Footwear_Units_Sold'
units_by_location = units_by_location.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Sort in descending order to see highest sales first
units_by_location = units_by_location.sort_values('Womens_Footwear_Units_Sold', ascending=False)

# Show the top 5 results
top_5_locations = units_by_location.head(5)

# Display the results
print("Top 5 locations for Women's Athletic Footwear sales by units:")
top_5_locations.head()

```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

# Create a pivot table
pivot_units = pd.pivot_table(
    womens_athletic_footwear,
    values='units_sold',    
    index=['retailer', 'region', 'state', 'city'], 
    aggfunc='sum'       
)

# Rename the column
pivot_units.columns = ['Womens_Footwear_Units_Sold']

# Reset the index to get a standard DataFrame
pivot_units = pivot_units.reset_index()

# Sort in descending order
pivot_units = pivot_units.sort_values('Womens_Footwear_Units_Sold', ascending=False)

# Show the top 5 results
top_5_pivot = pivot_units.head(5)

# Display the results
print("Top 5 locations for Women's Athletic Footwear sales by units")
top_5_pivot.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values

# Create a pivot table with invoice_date as index and total_sales as values
sales_by_date = pd.pivot_table(
    combined_df,
    values='total_sales',
    index=['invoice_date'],
    aggfunc='sum'
)

# Rename the column to "Total Sales"
sales_by_date = sales_by_date.rename(columns={'total_sales': 'Total Sales'})

# Display the table
print("Sales by Invoice Date:")
print(sales_by_date)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day
daily_sales = sales_by_date.resample('D').sum()

# Sort in descending order by "Total Sales"
daily_sales_sorted = daily_sales.sort_values('Total Sales', ascending=False)

# Display the result
print("Daily Sales (Sorted by Highest Sales):")
print(daily_sales_sorted)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = sales_by_date.resample('W').sum()

# Sort in descending order by "Total Sales"
weekly_sales_sorted = weekly_sales.sort_values('Total Sales', ascending=False)

# Display the result
print("Weekly Sales (Sorted by Highest Sales):")
print(weekly_sales_sorted)
```
